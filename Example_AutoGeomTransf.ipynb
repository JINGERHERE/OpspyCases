```python
import openseespy.opensees as ops
import opstool as opst
from ops_utilities.pre import AutoTransf
```

```python
ops.wipe()
ops.model('basic', '-ndm', 2, '-ndf', 3)

ops.node(1, 0., 0.)
ops.node(2, 0., 1.)
ops.node(3, 1., 1.)
ops.node(4, 1., 0.)

ops.geomTransf('Linear ', 100)
ops.geomTransf('PDelta  ', 101)
ops.geomTransf('Corotational', 102)

# element elasticBeamColumn $eleTag $iNode $jNode $A $E $Iz $transfTag
ops.element('elasticBeamColumn', 1, *(1, 2), 1., 1., 1., AutoTransf.ndm2())
ops.element('elasticBeamColumn', 2, *(2, 3), 1., 1., 1., AutoTransf.ndm2())
ops.element('elasticBeamColumn', 3, *(3, 4), 1., 1., 1., AutoTransf.ndm2())
```

```python
opst.vis.plotly.plot_model(show_local_axes=True)
```

```python
ops.getCrdTransfTags()
```

```python
ops.wipe()
ops.model('basic', '-ndm', 3, '-ndf', 6)

ops.node(1, 0., 0., 0.)
ops.node(2, 0., 1., 0.)
ops.node(3, 1., 1., 0.)
ops.node(4, 1., 0., 0.)

ops.node(5, 0., 0., 1.)
ops.node(6, 0., 1., 1.)
ops.node(7, 1., 1., 1.)
ops.node(8, 1., 0., 1.)

ops.fixZ(0., 1, 1, 1, 1, 1, 1)

# element elasticBeamColumn $eleTag $iNode $jNode $A $E $G $J $Iy $Iz $transfTag
ops.element('elasticBeamColumn', 1, *(1, 5), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(1, 5)))
ops.element('elasticBeamColumn', 2, *(2, 6), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(2, 6), 30))
ops.element('elasticBeamColumn', 3, *(3, 7), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(3, 7)))
ops.element('elasticBeamColumn', 4, *(4, 8), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(4, 8)))

ops.element('elasticBeamColumn', 5, *(5, 6), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(5, 6), 35))
ops.element('elasticBeamColumn', 6, *(6, 7), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(6, 7)))
ops.element('elasticBeamColumn', 7, *(7, 8), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(7, 8)))
ops.element('elasticBeamColumn', 8, *(8, 5), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(8, 5)))

ops.element('elasticBeamColumn', 9, *(2, 8), 1., 100., 10., 1., 1., 1., AutoTransf.ndm3(*(2, 8), 20))
```
